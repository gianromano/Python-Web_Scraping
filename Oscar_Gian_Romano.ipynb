```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:%C3%93scar_a_la_mejor_pel%C3%ADcula'
```

```python
def oscar_awards_statistics(url):

    response = requests.get(url)

    if response.status_code == 200:
        df_final = pd.DataFrame()
        soup = BeautifulSoup(response.content, 'html.parser')

        # List to collect DataFrames from each iteration
        dfs_to_append = []

        tables = soup.find_all("table")
        limit_tables = len(tables)

        for x in range(0, limit_tables):
            extracting = tables[x]

            header_list = extracting.find_all("th")
            limit_th = len(header_list)

            # hacer un for para th
            j = 0           
            for i in range(7,limit_th):
                    award_list = extracting.find_all("td")
                    limit_td = len(award_list)
                    oscar_year = [header_list[i].a.text]
                    # hacer un for para td
                    if j < limit_td:
                        a = j
                        b = j+1
                        c = j+2
                        d = j+3
                        e = j+4
                        j = e + 4
                        
                        movie_title = [award_list[a].a.text]
                        movie_studio = [award_list[b].text.rstrip('\n')]
                        movie_productor = [award_list[c].text.rstrip('\n')]
                        movie_director = [award_list[d].a.text]
                        movie_genre = [award_list[e].text.rstrip('\n')]
                                   
                        df_current = pd.DataFrame({
                        'Oscar Year': oscar_year,
                        'Movie Title' : movie_title,
                        'Movie Studio' : movie_studio,
                        'Movie Productor' : movie_productor,
                        'Movie Director' : movie_director,
                        'Movie Genre' : movie_genre
                        })
            
                        # Append the DataFrame to the list
                        dfs_to_append.append(df_current)

        # Concatenate all DataFrames in the list into a single DataFrame
        df_final = pd.concat(dfs_to_append,axis=0, ignore_index=True)

    return df_final
```

```python
df_final = oscar_awards_statistics(url)
#df_final.to_csv('oscar.csv', sep=',', header=True, index=False, encoding='utf-8')
df_final
```
