```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://es.wikipedia.org/wiki/Premios_Goya'
```

```python
def goya_awards(url):

    response = requests.get(url)

    if response.status_code == 200:
        df_final = pd.DataFrame()
        soup = BeautifulSoup(response.content, 'html.parser')

        # List to collect DataFrames from each iteration
        dfs_to_append = []

        table = soup.find_all("tbody")
        productores = table[2]
        datos = productores.findAll("td")
        limit_tables = len(datos)-3
        x = 0
        while x in range(0, limit_tables):
            if x < limit_tables:
                a = x
                b = x + 3
                x = b+3
                movie_year = [productores.findAll("td")[a].text.rstrip('\n')]
                movie_director = [productores.findAll("td")[b].text.strip('\n').replace('\n',' ')]
                                       
                df_current = pd.DataFrame({
                'Goya Year': movie_year,
                'Movie Director' : movie_director,
                })
                
                # Append the DataFrame to the list
                dfs_to_append.append(df_current)

        # Concatenate all DataFrames in the list into a single DataFrame
        df_final = pd.concat(dfs_to_append,axis=0, ignore_index=True)

    return df_final
    
```

```python
df_final = goya_awards(url)
df_final.to_csv('goya.csv',sep=',')
df_final
```
